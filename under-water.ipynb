```python
import os
import cv2
import glob

# ğŸ”§ Set this to your dataset root folder
dataset_root = "/content/under-water-object/URPC2020"

# âœ… List of dataset folders to check
splits = ["train", "valid", "test"]

bad_files = []

for split in splits:
    image_dir = os.path.join(dataset_root, split, "images")
    if not os.path.exists(image_dir):
        print(f"âš ï¸ Skipping {split}: no 'images' folder found.")
        continue

    print(f"\nğŸ” Checking images in: {image_dir}")
    image_files = glob.glob(os.path.join(image_dir, "*"))
    print(f"Found {len(image_files)} images.")

    for img_path in image_files:
        img = cv2.imread(img_path)
        if img is None:
            bad_files.append(img_path)
            print(f"âŒ Corrupted or unreadable: {img_path}")

# ğŸ§¹ Optionally delete bad files
if bad_files:
    print(f"\nğŸ§¹ Removing {len(bad_files)} bad images...")
    for f in bad_files:
        try:
            os.remove(f)
        except Exception as e:
            print(f"Error deleting {f}: {e}")
else:
    print("\nâœ… All images are readable!")
```

```python
import yaml, os

# check dataset folder
!ls /kaggle/input/under-water-object/URPC2020

DATASET_DIR = "/kaggle/input/under-water-object/URPC2020"

data_yaml = {
    "train": f"{DATASET_DIR}/train/images",
    "val": f"{DATASET_DIR}/valid/images",
    "test": f"{DATASET_DIR}/test/images",
    "nc": 4,   # number of classes in URPC2020
    "names": ["echinus", "holothurian", "scallop", "starfish"]
}

with open("/kaggle/input/waterde/data.yaml", "w") as f:
    yaml.dump(data_yaml, f, default_flow_style=False)

print(open("data.yaml").read())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # small model, you can change to yolov8m.pt

model.train(
    data="data.yaml",   # âœ… use the new yaml, not the one in /input/
    epochs=30,
    imgsz=640,
    batch=16,
    name="urpc2020_yolo"
)
```

```python
!pip install -q gradio

import gradio as gr
import cv2

# Load trained model from Kaggle output
trained_model = YOLO("/kaggle/input/waterde/best_model.pt")

def predict(image):
    results = trained_model.predict(source=image, save=False, conf=0.5)
    annotated_frame = results[0].plot()  # numpy image with bboxes
    return annotated_frame

iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="filepath"),
    outputs=gr.Image(type="numpy"),
    title="URPC2020 Underwater Object Detection",
    description="Upload an underwater image to detect echinus, holothurian, scallop, or starfish"
)

iface.launch(debug=True, share=True)  # share=True gives public link
```
